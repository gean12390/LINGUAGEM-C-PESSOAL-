**(Valor: 2,0 pts)** A fórmula de Bhaskara é uma fórmula matemática usada para resolver equações quadráticas, ou seja, equações do segundo grau da forma $ax^2+bx+c$. A fórmula é dada por:

$$ x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} $$

A parte dentro da raiz, $b^2-4ac^2-4ac$ ​ é conhecida como o discriminante (delta, Δ), que pode indica a natureza das raízes da equação:

* Se $Δ>0$, a equação tem duas raízes reais e distintas.
* Se $Δ=0$, a equação tem uma raiz real dupla (ou seja, duas raízes iguais).
* Se $Δ<0$, a equação não possui raízes reais.
